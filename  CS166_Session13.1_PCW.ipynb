## PageRank Networks 

```python
import networkx as nx 
import matplotlib.pyplot as plt 
import numpy as np 
import scipy.stats as sts 
```

```python
g = nx.erdos_renyi_graph(30, 0.05, directed = True, seed = 123)

nx.draw(g, pos = nx.kamada_kawai_layout(g))
```

```python
def random_surf(g, n):
    # Initialize nodes 
    g.nodes(data = 'pagerank', default = 0)
    g.nodes(data = 'visits', default = 0)
    
    alpha = 0.85 
    t = 0
    
    # Track values 
    
    # Choose initial surfer 
    surfer = np.random.choice(g.nodes)
    
    for i in range(n):
        t += 1

        neighbors = [n for n in g.neighbors(surfer)]
        
        if neighbors == []:
            surfer = np.random.choice(g.nodes)
            
        # Update 
        g.nodes[surfer]['visits'] += 1 
        g.nodes[surfer]['pagerank'] += 1/t 
        
        
        if np.random.uniform() < (alpha):
            surfer = np.random.choice(neighbors)
                        
        else:
            surfer = np.random.choice(g.nodes)
            
    

random_surf(g, 100)
```
